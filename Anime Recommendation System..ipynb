```python
#anime_id
#name
#genre
#type
anime = anime[['anime_id','name','genre','type']]
anime.head()
```

```python
anime['tags'] = anime['genre']+anime['type']
anime
```

```python
new_df = anime[['anime_id','name','tags']]
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.strip(','))
new_df
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=12017,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(anime):
    anime_index = new_df[new_df['name'] == anime].index[0]
    distances = similarity[anime_index]
    animes_List = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:10]
    for i in animes_List:
        print(new_df['name'].iloc[i[0]])
```

```python
recommend('Code Geass')
```
